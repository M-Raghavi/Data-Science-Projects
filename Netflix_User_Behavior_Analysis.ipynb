# ✅ Step 1: Import Libraries

```python
import numpy as np # linear algebra
import pandas as pd # data processing
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
%matplotlib inline
sns.set()
```

# 📥  Step 2: Load Dataset


```python
df=pd.read_csv('/content/netflix_titles.csv.zip')
df
```

# 🔍  Step 3: Basic Info and Missing Values


```python
# Basic info
print("Shape of dataset:", df.shape)
print("----------------------")
#information about dataset
print("\nDataset Info:")
print(df.info())
print("----------------------")
# Missing values
print("\nMissing Values:")
print(df.isnull().sum())
#print first five rows
print("--------------------")
print(df.head())
print("--------------------")
#summary of data
print("\nSummary of Data:")
print(df.describe())
```

# 🧼 Step 4: Data Cleaning



```python
# Drop rows with no title (title is a crucial field)
df = df.dropna(subset=['title'])

# Fill missing 'country' values with 'Unknown'
df['country'] = df['country'].fillna('Unknown')

# Fill missing 'cast' and 'director' with 'Not Specified'
df['cast'] = df['cast'].fillna('Not Specified')
df['director'] = df['director'].fillna('Not Specified')

# Fill missing 'rating' with 'Unknown'
df['rating'] = df['rating'].fillna('Unknown')

# Check that all missing values are handled
print("\nAfter Cleaning:")
print(df.isnull().sum())
```

# 📊 Step 5: Content Type Distribution (Bar Chart)



```python
# Bar chart to show the number of Movies vs TV Shows
plt.figure(figsize=(6,4))
type_counts = df['type'].value_counts()
plt.bar(type_counts.index, type_counts.values, color=['skyblue', 'lightgreen'])
plt.title('Distribution of Content Type')
plt.xlabel('Type')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

# 🌍 Step 6: Top 10 Countries Producing Content (Horizontal Bar)

```python
# Get top 10 countries
top_countries = df['country'].value_counts().head(10)

# Plot horizontal bar chart
plt.figure(figsize=(8, 5))
plt.barh(top_countries.index, top_countries.values, color='skyblue')
plt.title('Top 10 Countries with Most Content')
plt.xlabel('Number of Titles')
plt.tight_layout()
plt.show()
```

# 📅 Step 7: Trend of Content Over Years (Line Chart)





```python
# Convert 'date_added' to datetime objects
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Extract year from date
df['release_year'] = df['date_added'].dt.year

# Count titles added each year
yearly_trend = df['release_year'].value_counts().sort_index()

# Plot line chart
plt.figure(figsize=(10, 5))
plt.plot(yearly_trend.index, yearly_trend.values, marker='o', color='coral')
plt.title('Number of Titles Added Each Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

# ⌛ Step 8: Duration Distribution for Movies (Pie Chart)



```python
# Filter only movies and clean duration
movies = df[df['type'] == 'Movie'].copy()
movies['duration'] = movies['duration'].str.replace(' min', '').astype(str)
movies = movies[movies['duration'].str.isdigit()]
movies['duration'] = movies['duration'].astype(int)

# Categorize duration
bins = [0, 60, 90, 120, 180]
labels = ['<60 mins', '60-90 mins', '90-120 mins', '>120 mins']
movies['duration_category'] = pd.cut(movies['duration'], bins=bins, labels=labels, include_lowest=True)

# Pie chart
duration_dist = movies['duration_category'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(duration_dist, labels=duration_dist.index, autopct='%1.1f%%', colors=sns.color_palette('Set2'))
plt.title('Movie Duration Distribution')
plt.show()
```

# 🎬 Step 9: Most Frequent Directors (Bar Chart)

```python
# Plotting the top 10 most frequent directors (excluding 'Not Specified') based on number of titles they directed on Netflix

top_directors = df['director'].value_counts().drop('Not Specified').head(10)

plt.figure(figsize=(8, 5))
plt.barh(top_directors.index[::-1], top_directors.values[::-1], color='darkseagreen')
plt.title('Top 10 Most Frequent Directors')
plt.xlabel('Number of Titles')
plt.tight_layout()
plt.show()
```

# 🔥 Step 10: Content Added by Month and Year (Heatmap)



```python
# Extract month from 'date_added' and store in a new column 'month_added'
df['month_added'] = df['date_added'].dt.month

# Extract year from 'date_added' and store in a new column 'year_added'
df['year_added'] = df['date_added'].dt.year

# Create a pivot table to count the number of shows added per month and year
# index: month, columns: year, values: count of show_id
heatmap_data = df.pivot_table(index='month_added',
                              columns='year_added',
                              values='show_id',
                              aggfunc='count')

# Set figure size for better visibility
plt.figure(figsize=(12, 6))

# Create a heatmap using seaborn
# 'annot=True' adds the numbers inside each cell
# 'fmt=".0f"' ensures no decimal points are shown
# 'cmap' sets the color theme of the heatmap
sns.heatmap(heatmap_data, cmap='YlGnBu', linewidths=0.5, annot=True, fmt='.0f')

# Add chart title and axis labels
plt.title('Content Added by Month and Year')
plt.xlabel('Year')
plt.ylabel('Month')

# Display the plot
plt.show()
```

# 📌 Step 11: Word Cloud of Netflix Titles



```python
# Combine all non-null titles into a single string separated by spaces
title_text = ' '.join(df['title'].dropna())

# Set the figure size for the word cloud visualization
plt.figure(figsize=(12, 6))

# Import the WordCloud library
from wordcloud import WordCloud

# Generate a word cloud from the combined title text
# - Width/height defines the size of the word cloud
# - Background color is white
# - Colormap 'inferno' defines the color scheme of the words
wordcloud = WordCloud(width=1000, height=500, background_color='white', colormap='inferno').generate(title_text)

# Display the generated word cloud image
plt.imshow(wordcloud, interpolation='bilinear')

# Remove axis ticks and labels for a cleaner look
plt.axis('off')

# Add a title to the plot
plt.title('Most Common Words in Netflix Titles')

# Adjust layout to avoid overlap
plt.tight_layout()

# Show the plot
plt.show()
```

# 📊 Step 12: Animated Bar Chart of Top Netflix Genres by Year

```python
import plotly.express as px

# Convert 'date_added' to datetime if not already
df['date_added'] = pd.to_datetime(df['date_added'])

# Extract year and split genres
df['year_added'] = df['date_added'].dt.year
df['genre'] = df['listed_in'].str.split(', ').str[0]  # Take first genre if multiple

# Group by year and genre, count the number of titles
genre_year_df = df.groupby(['year_added', 'genre']).size().reset_index(name='count')

# Filter years with sufficient data (optional: from 2010 onward)
genre_year_df = genre_year_df[genre_year_df['year_added'] >= 2010]

# Get top 10 genres overall for filtering
top_genres = df['genre'].value_counts().head(10).index
genre_year_df = genre_year_df[genre_year_df['genre'].isin(top_genres)]

# Create an animated bar chart
fig = px.bar(genre_year_df,
             x='count',
             y='genre',
             color='genre',
             animation_frame='year_added',
             orientation='h',
             title='🎬 Top Netflix Genres by Year (Animated)',
             labels={'count': 'Number of Titles', 'genre': 'Genre'},
             template='plotly_white',
             height=500)

# Invert y-axis to show highest bars on top
fig.update_layout(yaxis={'categoryorder': 'total ascending'})

# Show the plot
fig.show()
```
